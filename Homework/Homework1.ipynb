```python
# 1. Create a line chart to visualize the distribution of the 'capitalLong' feature.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Ensure the 'capitalLong' column exists
if 'capitalLong' in data.columns:
    # Sort the data by 'capitalLong' if required
    data_sorted = data.sort_values(by='capitalLong').reset_index(drop=True)

    # Plotting the line chart
    plt.figure(figsize=(12, 6))
    plt.plot(data_sorted.index, data_sorted['capitalLong'], color='b', marker='o', linestyle='-', linewidth=1.5)

    # Adding labels and title
    plt.xlabel('Index')
    plt.ylabel('Capital Longitude')
    plt.title('Distribution of Capital Longitude')

    # Display the plot
    plt.show()
```

```python
# 2. Generate a bar chart to show the frequency of "type"(spam and nonspam) in the dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Check if 'type' column exists
if 'type' in data.columns:
    # Count the frequency of each type
    type_counts = data['type'].value_counts()

    # Plotting the bar chart
    plt.figure(figsize=(12, 6))
    type_counts.plot(kind='bar', color=['red', 'blue'])

    # Adding labels and title
    plt.xlabel('Type')
    plt.ylabel('Frequency')
    plt.title('Frequency of Spam and Nonspam Emails')

    # Display the plot
    plt.show()
```

```python
# 3. Create a scatter plot to explore the relationship between 'num3d' and 'num000'.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Check if both 'num3d' and 'num000' columns exist
if 'num3d' in data.columns and 'num000' in data.columns:
    # Plotting the scatter plot
    plt.figure(figsize=(12, 6))
    plt.scatter(data['num3d'], data['num000'], color='purple', alpha=0.6, edgecolor='k')

    # Adding labels and title
    plt.xlabel('num3d')
    plt.ylabel('num000')
    plt.title('Relationship between num3d and num000')

    # Display the plot
    plt.show()
```

```python
# 4. Visualize the distribution of 'money' using a histogram.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Ensure the 'money' column exists
if 'money' in data.columns:
    # Plotting the histogram
    plt.figure(figsize=(12, 10))
    plt.hist(data['money'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)

    # Adding labels and title
    plt.xlabel('Money')
    plt.ylabel('Frequency')
    plt.title('Distribution of Money')

    # Display the plot
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
# 5. Create a pie chart to represent the proportion of spam and nonspam emails in the dataset.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Check if 'type' column exists
if 'type' in data.columns:
    # Count the frequency of each type (spam and nonspam)
    type_counts = data['type'].value_counts()

    # Plotting the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue'])

    # Adding title
    plt.title('Proportion of Spam and Nonspam Emails')

    # Display the plot
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

```python
# 6. Generate a table chart to display the summary statistics for the 'capitalAve' feature.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Ensure the 'capitalAve' column exists
if 'capitalAve' in data.columns:
    # Calculate summary statistics
    summary_stats = data['capitalAve'].describe()

    # Create a DataFrame for displaying as a table
    summary_df = pd.DataFrame(summary_stats)

    # Add index labels to the DataFrame
    summary_df.index = ['Count', 'Mean', 'Standard Deviation', 'Min', '25%', 'Median (50%)', '75%', 'Max']

    # Create a new DataFrame with the descriptions as the first column
    description = [
        'Number of observations',
        'Average value of capitalAve',
        'Measure of dispersion from the mean',
        'Smallest value in capitalAve',
        'First quartile (25th percentile)',
        'Middle value of capitalAve',
        'Third quartile (75th percentile)',
        'Largest value in capitalAve'
    ]
    
    # Adding the description as the first column
    summary_df['Description'] = description

    # Rearranging the columns to have Description first
    summary_df = summary_df[['Description', summary_df.columns[0]]]  # First two columns
    summary_df.columns = ['Description', 'Value']  # Renaming columns for clarity

    # Plotting the table
    plt.figure(figsize=(12, 6))
    plt.axis('tight')
    plt.axis('off')  # Turn off the axes

    # Create a table
    table = plt.table(cellText=summary_df.values,
                      colLabels=summary_df.columns,
                      cellLoc='center',
                      loc='center')

    # Adding title
    plt.title('Summary Statistics for capitalAve')

    # Display the plot
    plt.show()
```

```python
# 7. Calculate the mean, median, and standard deviation of the 'our' feature
import pandas as pd

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Ensure the 'our' column exists
if 'our' in data.columns:
    # Calculate mean, median, and standard deviation
    mean_value = data['our'].mean()
    median_value = data['our'].median()
    std_dev_value = data['our'].std()

    # Print the results
    print(f"Mean of 'our': {mean_value}")
    print(f"Median of 'our': {median_value}")
    print(f"Standard Deviation of 'our': {std_dev_value}")
```

```python
# 8. Determine the minimum and maximum values of 'hp'
import pandas as pd

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Ensure the 'hp' column exists
if 'hp' in data.columns:
    # Calculate minimum and maximum values
    min_value = data['hp'].min()
    max_value = data['hp'].max()

    # Print the results
    print(f"Minimum value of 'hp': {min_value}")
    print(f"Maximum value of 'hp': {max_value}")
```

```python
# 9. Calculate the 25th, 50th, and 75th percentiles for 'telnet'
import pandas as pd

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Ensure the 'telnet' column exists
if 'telnet' in data.columns:
    # Calculate the 25th, 50th, and 75th percentiles
    percentiles = data['telnet'].quantile([0.25, 0.50, 0.75])

    # Print the results
    print(f"25th percentile of 'telnet': {percentiles[0.25]}")
    print(f"50th percentile (median) of 'telnet': {percentiles[0.50]}")
    print(f"75th percentile of 'telnet': {percentiles[0.75]}")
```

```python
# 10.  Find the mode of the 'you' feature.
import pandas as pd

# Load the dataset
data = pd.read_csv('Dataset for HW1.csv')

# Ensure the 'you' column exists
if 'you' in data.columns:
    # Calculate the mode
    mode_value = data['you'].mode()

    # Print the results
    if not mode_value.empty:
        print(f"Mode of 'you': {mode_value[0]}")  # Mode can have multiple values, take the first one
    else:
        print("No mode found for 'you'.")
else:
    print("The 'you' column was not found in the dataset.")
```
