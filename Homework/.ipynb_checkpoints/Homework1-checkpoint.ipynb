```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
data = pd.read_csv("Dataset for HW1.csv")

# Set default figure size
plt.rcParams['figure.figsize'] = (14, 10)
```

```python
# 1. Create a line chart to visualize the distribution of the 'capitalLong' feature.

# Plot each point with markers and connect them with lines
plt.plot(data.index, data['capitalLong'])

# Add labels and title
plt.title('Line Chart of the dispersion of capitalLong')
plt.xlabel('Index')
plt.ylabel('capitalLong')

# Display the graph
plt.show()
```

```python
# 2. Generate a bar chart to show the frequency of "type"(spam and nonspam) in the dataset

# Plot the Data
plt.figure()
sns.countplot(x='type', data=data, hue='type', palette={'spam': 'red', 'nonspam': 'blue'}, legend=False)

# Add Labels and Title
plt.title('Frequency of Spam and Non-Spam')
plt.xlabel('Type')
plt.ylabel('Frequency')

#Display the Graph
plt.show()
```

```python
# 3. Create a scatter plot to explore the relationship between 'num3d' and 'num000'.

# Plot the Data
plt.figure()
sns.scatterplot(x='num3d', y='num000', data=data)

# Add Labels and Title
plt.title('Relationship between num3d and num000')
plt.xlabel('num3d')
plt.ylabel('num000')

#Display the Graph
plt.show()
```

```python
# 4. Visualize the distribution of 'money' using a histogram.

#Plot the Data
plt.figure()
sns.histplot(data['money'], kde=False)

# Adding labels and title
plt.title('Distribution of Money')
plt.xlabel('Money')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# 5. Create a pie chart to represent the proportion of spam and nonspam emails in the dataset.

# Plotting the pie chart
plt.figure()
data['type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['blue','red'])

# Adding title
plt.title('Proportion of Spam and Non-Spam Emails')
plt.ylabel('')  # To hide the y-axis label

# Display the plot
plt.show()
```

```python
# 6. Generate a table chart to display the summary statistics for the 'capitalAve' feature.

capitalAve_summary = data['capitalAve'].describe()
print("Summary Statistics for capitalAve:")
print(capitalAve_summary.to_frame().T)
```

```python
# 7. Calculate the mean, median, and standard deviation of the 'our' feature

# Calculate mean, median, and standard deviation
mean_value = data['our'].mean()
median_value = data['our'].median()
std_dev_value = data['our'].std()

# Print the results
print(f"Mean of 'our': {mean_value}")
print(f"Median of 'our': {median_value}")
print(f"Standard Deviation of 'our': {std_dev_value}")
```

```python
# 9. Calculate the 25th, 50th, and 75th percentiles for 'telnet'

# Calculate the 25th, 50th, and 75th percentiles
percentiles_telnet = np.percentile(data['telnet'], [25, 50, 75])

# Print the results
print("\n25th percentile of 'telnet':", percentiles_telnet[0])
print("50th percentile (median) of 'telnet':", percentiles_telnet[1])
print("75th percentile of 'telnet':", percentiles_telnet[2])
```

```python
# 10.  Find the mode of the 'you' feature.

# Calculate the mode
mode_you = data['you'].mode()

# Print the results
print("\nMode of 'you':", mode_you)
```

```python
# 11. Calculate the interquartile range (IQR) for 'business'.

# Calculate the interquartile range (IQR)
Q1 = data['business'].quantile(0.25)
Q3 = data['business'].quantile(0.75)
IQR = Q3 - Q1

# Print the Results
print("\nInterquartile Range (IQR) for 'business':", IQR)
```

```python
# 12. Compute the skewness and kurtosis of the 'credit' feature.

# Compute the skewness and kurtosis
skewness_credit = data['credit'].skew()
kurtosis_credit = data['credit'].kurtosis()

# Print the Results
print("\nSkewness of 'credit':", skewness_credit)
print("Kurtosis of 'credit':", kurtosis_credit)
```

```python
17. Identify the two features with the highest positive correlation and the two with the highest
negative correlation.
```

```python
import pandas as pd

data = pd.read_csv('spambase.csv')

# Select only numeric columns
numeric_data = data.select_dtypes(include='number')

# Generate the correlation matrix
correlation_matrix = numeric_data.corr()

# Get the absolute values of the correlation matrix
abs_corr_matrix = correlation_matrix.abs()

# Mask the diagonal (self-correlation) by setting it to zero
import numpy as np
np.fill_diagonal(abs_corr_matrix.values, 0)

# Find the two features with the highest positive correlation
highest_positive_corr = abs_corr_matrix.unstack().sort_values(ascending=False)
top_positive = highest_positive_corr.head(2)

# Find the two features with the highest negative correlation
highest_negative_corr = correlation_matrix.unstack().sort_values()
top_negative = highest_negative_corr.head(2)

# Display the results
print("Two features with the highest positive correlation:")
print(top_positive)

print("\nTwo features with the highest negative correlation:")
print(top_negative)
```

```python
18. Visualize the correlation matrix using a heatmap
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('spambase.csv')

# Select only numeric columns
numeric_data = data.select_dtypes(include='number')

# Generate the correlation matrix for numeric features
correlation_matrix = numeric_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,         # Show correlation values
    fmt=".2f",         # Format the numbers to two decimal places
    cmap='coolwarm',   # Color map
    square=True,       # Make cells square-shaped
    cbar_kws={"shrink": .8},  # Shrink color bar
    linewidths=0.5,   # Lines between cells
    linecolor='white'  # Color of the lines
)

# Set the title
plt.title('Correlation Matrix of Numeric Features', fontsize=16)

# Show the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

```python
23. Describe the characteristics of data that make a bar chart suitable for visualization.

Bar charts are an easy and distinct way to be able to visualize data. We can use bar charts to view categorical variables
(such as hair or eye color) or discrete data that is countable and can only be measured in whole numbers (e.g pages in book or students
in a classroom).                                                                                                        
```

```python
24. Explain when it is appropriate to use a scatter plot for data exploration

Scatter plots are appropriate to use when we want to see trends in data or see the association between data that is continious.
They are also useful for visualizing outliers that are significantly different from other data points.
```

```python
25. Discuss the advantages and disadvantages of using a pie chart for categorical data

Advantages: Pie charts are visually easy to understand how data contributes to the "whole" of a pie. They allow us to quickly compare
different catergories of data by looking at the "slice" of each category in the pie chart, which makes them more appealing when presenting 
data to an audience.

Disadvantages: Pie charts are not best when wanting to display a large amount of catergories and work best when there are 7 or fewer
categories. They can make it harder to compare categories especially if there are small differences which may lead to misinterpretation of data.
```

```python
26. When would you choose a line chart over a histogram for data visualization?

Line charts are best used over a histogram when wanting to display trends of continious variables over a period time, such as the daily 
temperature or fluctuation of currency exchange rates.
```

```python
27. Describe a scenario where a table chart is more informative than other chart types

Table charts are useful when we want to compare multiple data points and they allow for many different categories.
A scenario where this would be imformative is when a company wants to compile an inventory list. A table chart could display the different 
products along with the associated cost, the quantity of items available.
```

```python
28. Explain the key insights you can gain from a histogram.

Histograms show the distribution of data and if it is normally distributed or not. We can also use them to identify potential outliters
or patterns in the data. They are also useful for seeing how data is distrubuted across ranges of intervals.
```
